#### Step 2: อ่านไฟล์ csv โดยใช้ pandas โดยเก็บไว้ในตัวแปร df

```python
import pandas as pd

df = pd.read_csv('datasets/aapl.csv')

# Convert the 'Date' column to a datetime object
df['Date'] = pd.to_datetime(df['Date'])

# filter only 2008 year
df_2008 = df[(df['Date'].dt.year == 2008)]

# Group the data by month and sum the 'Volume' column
monthly_volume_2008 = df_2008.groupby(df_2008['Date'].dt.month)['Volume'].sum()

# Convert the monthly volume to a DataFrame
monthly_volume_df = pd.DataFrame({
    'Month': monthly_volume_2008.index, 
    'Volume': monthly_volume_2008.values
    }
)

monthly_volume_df
# df = df[df['Date'].str.contains('2008')]

# filter only 2008 year and volume > 50 million
# df = df[(df['Date'].str.contains('2008')) & (df['Volume'] > 50000000)]

# filter only 2008 grop by month sum volume
# df = df[df['Date'].str.contains('2008')].groupby('Date').sum().reset_index()

# df
```

#### Step 3: Plot Graph โดยใช้ matplotlib

```python
import matplotlib.pyplot as plt

# plot line chart date (x) vs volume (y)
plt.plot(monthly_volume_df['Month'], monthly_volume_df['Volume'])

# show the chart
plt.show()
```

```python
# plot line chart date (x) vs volume (y)
plt.bar(monthly_volume_df['Month'], monthly_volume_df['Volume'])

# show the chart
plt.show()
```
